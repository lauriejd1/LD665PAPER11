```python
# Section 3, Task 3

# Below is a slightly different for loop approach, which uses the melt variabel from the Arolla dataset given above.
# For this task, we just want you to look at the code, and comment the lines to work out what the code is doing.
# At the end of this, we ask you to summarise the for-loop in max. 200 words.

# Extract the 'Melt' column from the DataFrame to work with the melt data
melt = df['Melt']

# Define periods of interest, each period is a tuple (start_day, end_day)
periods = [(150, 169), (170, 189), (190, 209)]

# Iterate over the periods defined above
for start, end in periods:
    # Create a boolean array 'I' that identifies whether 'Day' is between the 'start' and 'end' of each period
    I = (Day >= start) & (Day <= end)

    # Use this boolean array to filter the 'Melt' data, extracting only the melt values for the current period
    Melt_I = melt[I]

    # Calculate the mean melt for the current period using numpy's mean function
    Mean_Melt = np.mean(Melt_I)

 # Print the result, showing the mean melt for the current period
    print(f"The mean melt at Glacier d'Arolla between days {start}-{end} was: {Mean_Melt:.2} cm")

# Write a 200 words summary of what this for-loop achieved.
#This for-loop calculates the mean melt for three specific periods at Glacier d'Arolla using the Melt data. It first defines three periods of interest: (150, 169), (170, 189), and (190, 209). For each period, the loop does the following:
#It generates a boolean array I, where True values correspond to days within the current period (i.e., the Day variable falls between the start and end days of the period).
#It then uses this boolean array to filter the Melt data, extracting only the melt values for the current period (Melt_I).
#The loop calculates the average melt for the period using np.mean().
#Finally, it prints the mean melt for each period with a message formatted to show the start and end days of the period along with the calculated mean.'''

# Can you think of another way you could utilise this 'start, end' functionality to analyse time series data? Please write a 100 word description of this.
#The 'start, end' functionality can be used to perform various types of time-based analyses. For instance, you could use this approach to calculate seasonal averages or detect anomalies. By defining time ranges corresponding to different seasons (e.g., winter, spring, etc.), you could compute the average values for each season and identify trends or changes across multiple years. Additionally, this functionality can be applied to compare time intervals with specific weather events or anomalies, such as extreme temperature changes, by adjusting the time periods accordingly. This method offers flexibility for temporal data analysis in many contexts.
```
